```python
#忽略警告提示
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
#导入数据
train=pd.read_csv('/Users/yezhibin/Desktop/train.csv')
#测试数据
test=pd.read_csv('/Users/yezhibin/Desktop/test.csv')

print('训练数据集：',train.shape,'测试训练集：',test.shape)
```

```python
#数据处理
full['Age']=full['Age'].fillna(full['Age'].mean())
full['Fare']=full['Fare'].fillna(full['Fare'].mean())
print('处理后：')
full.info()
```

```python
sex_mapDict={'male':1,
             'female':0}
full['Sex']=full['Sex'].map(sex_mapDict)
full.head()
```

```python
full['Embarked'].head()
```

```python
embarkedDf=pd.DataFrame()

embarkedDf=pd.get_dummies(full['Embarked'],prefix='Embarked')
embarkedDf.head()
```

```python
#添加one-hot编码产生的虚拟变量（dummy variables）到泰坦尼克号数据集full
full = pd.concat([full,embarkedDf],axis=1)

'''
因为已经使用登船港口(Embarked)进行了one-hot编码产生了它的虚拟变量（dummy variables）
所以这里把登船港口(Embarked)删掉
'''
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
full=pd.concat([full,embarkedDf],axis=1)

full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
pclassDf=pd.DataFrame()

pclassDf=pd.get_dummies(full['Pclass'],prefix='Pclass')
pclassDf.head()
```

```python
full=pd.concat([full,pclassDf],axis=1)

full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
def getTitle(name):
    str1=name.split(',')[1]
    str2=str1.split('.')[0]
    str3=str2.strip()
    
    return str3
```

```python
titleDf=pd.DataFrame()
titleDf['Title']=full['Name'].map(getTitle)
titleDf.head()
```

```python
title_mapDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }

titleDf['title']=titleDf['Title'].map(title_mapDict)

titleDf=pd.get_dummies(titleDf['Title'])
titleDf.head()
```

```python
full=pd.concat([full,titleDf],axis=1)
full.drop('Name',axis=1,inplace=True)
full.head()
```
